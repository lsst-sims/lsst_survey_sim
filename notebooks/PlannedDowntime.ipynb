```python
import numpy as np
import pandas as pd
from astropy.time import Time, TimeDelta

from rubin_scheduler.site_models import ScheduledDowntimeData 
```

```python
tstart = Time("2026-01-01T12:00:00")
tend = tstart + TimeDelta(365.25*10, format='jd')
sd = ScheduledDowntimeData(tstart)
```

```python
dd = []
for dt in sd.downtime:
    dd.append([dt['start'].iso, dt['end'].iso, dt['end'] - dt['start'], dt['activity']])
dd = pd.DataFrame(dd, columns=["start", "end", "length (days)", "activity"])
dd.query("start < @tend")
```
